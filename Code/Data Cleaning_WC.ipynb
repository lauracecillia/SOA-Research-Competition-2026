# Data Cleaning

```python
import pandas as pd
import numpy as np
import re

# ─────────────────────────────────────────
# HELPER FUNCTIONS
# ─────────────────────────────────────────

def clean_numeric_range(df, col, low, high):
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df.loc[(df[col] < low) | (df[col] > high), col] = np.nan

def clean_discrete(df, col, valid_values):
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df.loc[~df[col].isin(valid_values), col] = np.nan

# ─────────────────────────────────────────
# CATEGORICAL STANDARDIZERS
# ─────────────────────────────────────────

def standardize_policy_id(val):
    """Format: WC-EPS-00001, WC-HEL-00001, WC-ZET-00001"""
    if pd.isna(val): return np.nan
    match = re.match(r'^(WC-(EPS|HEL|ZET)-\d+)$', str(val).strip(), re.IGNORECASE)
    return match.group(1).upper() if match else np.nan

def standardize_worker_id(val):
    """Format: W-0001"""
    if pd.isna(val): return np.nan
    match = re.match(r'^(W-\d+)$', str(val).strip(), re.IGNORECASE)
    return match.group(1).upper() if match else np.nan

def standardize_claim_id(val):
    """Format: C-0001286"""
    if pd.isna(val): return np.nan
    match = re.match(r'^(C-\d+)$', str(val).strip(), re.IGNORECASE)
    return match.group(1).upper() if match else np.nan

def standardize_solar_system(val):
    if pd.isna(val): return np.nan
    v = str(val).strip().lower()
    if 'helionis' in v: return 'Helionis Cluster'
    if 'epsilon' in v: return 'Epsilon'
    if 'zeta' in v: return 'Zeta'
    return np.nan

def standardize_station_id(val):
    if pd.isna(val): return np.nan
    match = re.match(r'^([A-Za-z][0-9]+)', str(val).strip())
    return match.group(1).upper() if match else np.nan

def standardize_occupation(val):
    valid = [
        'Administrator', 'Manager', 'Executive', 'Scientist',
        'Engineer', 'Operator', 'Planetary Operations', 'Drill Operator',
        'Safety Officer', 'Spacecraft Operator', 'Technology Officer',
        'Maintenance Staff'
    ]
    if pd.isna(val): return np.nan
    v = str(val).strip()
    # Buang suffix _???XXXX dulu
    v_clean = re.sub(r'_\?\?\?\d+$', '', v).strip()
    if v_clean in valid:
        return v_clean
    return np.nan

def standardize_employment_type(val):
    if pd.isna(val): return np.nan
    v = str(val).strip().lower()
    if 'full' in v: return 'Full time'
    if 'contract' in v: return 'Contract'
    return np.nan

def standardize_injury_type(val):
    valid = {
        'Cut laceration': 'Cut laceration',
        'Sprain, strain': 'Sprain, strain',
        'Stress': 'Stress',
        'Burns': 'Burns',
        'Psychological': 'Psychological',
        'Other': 'Other',
        'Amputation': 'Amputation',
        'Fracture': 'Fracture',
        'Crush': 'Crush',
        'Contusion': 'Contusion',
        'Dislocation': 'Dislocation',
        'Poisoning': 'Poisoning'
    }
    if pd.isna(val): return np.nan
    # Buang suffix _???XXXX
    v = re.sub(r'_\?\?\?\d+$', '', str(val).strip())
    if v in valid:
        return valid[v]
    return np.nan

# Apply ulang
df_sev['injury_type'] = df_sev['injury_type'].apply(standardize_injury_type)

def standardize_injury_cause(val):
    valid = [
        'Stress/strain', 'Vehicle accident', 'Exposure',
        'Overexertion', 'Caught in machine', 'Other',
        'Violence', 'Fall', 'Equipment failure',
        'Chemical', 'Explosion', 'Electrical', 'Struck by object'
    ]
    if pd.isna(val): return np.nan
    v = re.sub(r'_\?\?\?\d+$', '', str(val).strip())
    for item in valid:
        if item.lower() in v.lower():
            return item
    return np.nan

# ─────────────────────────────────────────
# CLEAN df_freq (17 variabel)
# ─────────────────────────────────────────

# 1. policy_id
df_freq['policy_id'] = df_freq['policy_id'].apply(standardize_policy_id)

# 2. worker_id
df_freq['worker_id'] = df_freq['worker_id'].apply(standardize_worker_id)

# 3. solar_system
df_freq['solar_system'] = df_freq['solar_system'].apply(standardize_solar_system)

# 4. station_id
df_freq['station_id'] = df_freq['station_id'].apply(standardize_station_id)

# 5. occupation
df_freq['occupation'] = df_freq['occupation'].apply(standardize_occupation)

# 6. employment_type
df_freq['employment_type'] = df_freq['employment_type'].apply(standardize_employment_type)

# 7. experience_yrs
clean_numeric_range(df_freq, 'experience_yrs', 0.2, 40)

# 8. accident_history_flag
clean_discrete(df_freq, 'accident_history_flag', [0, 1])

# 9. psych_stress_index
clean_discrete(df_freq, 'psych_stress_index', [1, 2, 3, 4, 5])

# 10. hours_per_week
clean_discrete(df_freq, 'hours_per_week', [20, 25, 30, 35, 40])

# 11. supervision_level
clean_numeric_range(df_freq, 'supervision_level', 0, 1)

# 12. gravity_level
clean_numeric_range(df_freq, 'gravity_level', 0.75, 1.50)

# 13. safety_training_index
clean_discrete(df_freq, 'safety_training_index', [1, 2, 3, 4, 5])

# 14. protective_gear_quality
clean_discrete(df_freq, 'protective_gear_quality', [1, 2, 3, 4, 5])

# 15. base_salary
clean_numeric_range(df_freq, 'base_salary', 20000, 130000)

# 16. exposure
clean_numeric_range(df_freq, 'exposure', 0, 1)

# 17. claim_count
clean_numeric_range(df_freq, 'claim_count', 0, 2)


# ─────────────────────────────────────────
# CLEAN df_sev (22 variabel)
# ─────────────────────────────────────────

# 1. claim_id
df_sev['claim_id'] = df_sev['claim_id'].apply(standardize_claim_id)

# 2. claim_seq
df_sev['claim_seq'] = pd.to_numeric(df_sev['claim_seq'], errors='coerce')
df_sev.loc[df_sev['claim_seq'] <= 0, 'claim_seq'] = np.nan

# 3. policy_id
df_sev['policy_id'] = df_sev['policy_id'].apply(standardize_policy_id)

# 4. worker_id
df_sev['worker_id'] = df_sev['worker_id'].apply(standardize_worker_id)

# 5. solar_system
df_sev['solar_system'] = df_sev['solar_system'].apply(standardize_solar_system)

# 6. station_id
df_sev['station_id'] = df_sev['station_id'].apply(standardize_station_id)

# 7. occupation
df_sev['occupation'] = df_sev['occupation'].apply(standardize_occupation)

# 8. employment_type
df_sev['employment_type'] = df_sev['employment_type'].apply(standardize_employment_type)

# 9. experience_yrs
clean_numeric_range(df_sev, 'experience_yrs', 0.2, 40)

# 10. accident_history_flag
clean_discrete(df_sev, 'accident_history_flag', [0, 1])

# 11. psych_stress_index
clean_discrete(df_sev, 'psych_stress_index', [1, 2, 3, 4, 5])

# 12. hours_per_week
clean_discrete(df_sev, 'hours_per_week', [20, 25, 30, 35, 40])

# 13. supervision_level
clean_numeric_range(df_sev, 'supervision_level', 0, 1)

# 14. gravity_level
clean_numeric_range(df_sev, 'gravity_level', 0.75, 1.50)

# 15. safety_training_index
clean_discrete(df_sev, 'safety_training_index', [1, 2, 3, 4, 5])

# 16. protective_gear_quality
clean_discrete(df_sev, 'protective_gear_quality', [1, 2, 3, 4, 5])

# 17. base_salary
clean_numeric_range(df_sev, 'base_salary', 20000, 130000)

# 18. exposure
clean_numeric_range(df_sev, 'exposure', 0, 1)

# 19. injury_type
df_sev['injury_type'] = df_sev['injury_type'].apply(standardize_injury_type)

# 20. injury_cause
df_sev['injury_cause'] = df_sev['injury_cause'].apply(standardize_injury_cause)

# 21. claim_length
clean_numeric_range(df_sev, 'claim_length', 3, 1000)

# 22. claim_amount
clean_numeric_range(df_sev, 'claim_amount', 5, 170000)
```

# MICE Imputation and Crossfill

```python
import pandas as pd
import numpy as np
import openpyxl

# Load Excel file
file_path = "/content/workcomp_match.xlsx"
print(file_path)
```

```python
new_freq = pd.read_excel(file_path, sheet_name="freq")
new_sev = pd.read_excel(file_path, sheet_name="sev")
```

```python
# ─────────────────────────────────────────
# CEK & FILL MISSING VIA LOOKUP ANTAR SHEET
# Key: policy_id + worker_id + station_id + solar_system
# ─────────────────────────────────────────

lookup_key = ['policy_id', 'worker_id', 'station_id', 'solar_system']

check_cols_sev  = [c for c in new_sev.columns  if c not in lookup_key + ['claim_id', 'claim_seq']]
check_cols_freq = [c for c in new_freq.columns if c not in lookup_key + ['claim_count']]
shared = [c for c in check_cols_sev if c in check_cols_freq]
print("Kolom crossfill:", shared)

def build_lookup(df, key_cols, val_cols):
    return (
        df.dropna(subset=key_cols)
          .groupby(key_cols)[val_cols]
          .first()
          .reset_index()
    )

lookup_from_freq = build_lookup(new_freq, lookup_key, shared)
lookup_from_sev  = build_lookup(new_sev,  lookup_key, shared)

def crossfill(df_target, lookup, key_cols, cols):
    """Fill NaN di df_target dari lookup berdasarkan key_cols."""
    df = df_target.copy()
    merged = df.merge(lookup, on=key_cols, how='left', suffixes=('', '_fill'))
    for col in cols:
        fill_col = f'{col}_fill'
        if fill_col not in merged.columns:
            continue
        filled = df[col].isna() & merged[fill_col].notna()
        if filled.sum() > 0:
            df.loc[filled, col] = merged.loc[filled, fill_col].values
            print(f"  {col}: {filled.sum()} terisi")
    return df

print("\n=== Fill new_sev dari new_freq ===")
new_sev = crossfill(new_sev, lookup_from_freq, lookup_key, shared)

print("\n=== Fill new_freq dari new_sev ===")
new_freq = crossfill(new_freq, lookup_from_sev, lookup_key, shared)

print("\n=== RINGKASAN SISA MISSING ===")
print("new_sev:")
sev_miss = new_sev[shared].isnull().sum()
print(sev_miss[sev_miss > 0])
print("\nnew_freq:")
freq_miss = new_freq[shared].isnull().sum()
print(freq_miss[freq_miss > 0])
```

```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.preprocessing import OrdinalEncoder

# ── 1. SETUP ──────────────────────────────────────────────────────────────────
id_cols          = ['policy_id', 'worker_id']
continuous_cols  = ['experience_yrs', 'supervision_level', 'gravity_level', 'base_salary']
ordinal_cols     = ['psych_stress_index', 'safety_training_index', 'protective_gear_quality']
binary_cols      = ['accident_history_flag']
discrete_cols    = ['hours_per_week']
nominal_cols     = ['solar_system', 'station_id', 'occupation', 'employment_type']

all_impute_cols  = continuous_cols + ordinal_cols + binary_cols + discrete_cols + nominal_cols

# ── 2. ENCODE NOMINAL CATS ────────────────────────────────────────────────────
df_imp = new_freq.drop(columns=id_cols).copy()

enc = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
df_imp[nominal_cols] = enc.fit_transform(df_imp[nominal_cols])

# Semua ke float64
for col in df_imp.columns:
    df_imp[col] = pd.to_numeric(df_imp[col], errors='coerce').astype('float64')

print("Dtypes OK:", df_imp.dtypes.unique())
print("Inf check:", np.isinf(df_imp.values).sum())
print("Missing before:", df_imp.isnull().sum().sum())

# ── 3. ITERATIVE IMPUTER ──────────────────────────────────────────────────────
from sklearn.linear_model import BayesianRidge

imputer = IterativeImputer(
    estimator=BayesianRidge(),  # jauh lebih cepat dari RF
    max_iter=20,
    random_state=42,
    verbose=2
)

imputed_array = imputer.fit_transform(df_imp)
df_imputed    = pd.DataFrame(imputed_array, columns=df_imp.columns, index=df_imp.index)

print("\nMissing after:", df_imputed.isnull().sum().sum())

# ── 4. POST-PROCESSING ────────────────────────────────────────────────────────

# Ordinal → round & clip ke 1–5
for col in ordinal_cols:
    df_imputed[col] = df_imputed[col].round().astype(int).clip(1, 5)

# Binary → round & clip ke 0–1
for col in binary_cols:
    df_imputed[col] = df_imputed[col].round().astype(int).clip(0, 1)

# hours_per_week → snap ke nilai valid
valid_hours = np.array([20, 25, 30, 35, 40])
df_imputed['hours_per_week'] = df_imputed['hours_per_week'].apply(
    lambda x: valid_hours[np.argmin(np.abs(valid_hours - x))]
)

# Decode nominal cats balik ke label asli
df_imputed[nominal_cols] = df_imputed[nominal_cols].round().astype(int)
df_imputed[nominal_cols] = enc.inverse_transform(
    df_imputed[nominal_cols].clip(
        0, [len(c) - 1 for c in enc.categories_]
    )
)

# Kembalikan ID cols
df_imputed[id_cols] = new_freq[id_cols].values

# ── 5. VALIDASI ───────────────────────────────────────────────────────────────
print("\n=== VALIDASI ===")
print("Missing:", df_imputed.isnull().sum().sum())

print("\nNumerik — mean shift:")
for col in continuous_cols:
    before = new_freq[col].mean()
    after  = df_imputed[col].mean()
    pct    = abs(after - before) / before * 100
    print(f"  {col:30s} | before: {before:.3f} | after: {after:.3f} | shift: {pct:.2f}%")

print("\nOrdinal — distribusi:")
for col in ordinal_cols:
    comp = pd.DataFrame({
        'before': new_freq[col].value_counts(normalize=True).sort_index(),
        'after' : df_imputed[col].value_counts(normalize=True).sort_index()
    })
    print(f"\n  {col}\n", comp.round(4))

print("\nhours_per_week:")
print(df_imputed['hours_per_week'].value_counts().sort_index())
```

```python
# Kolom nominal sudah string — skip re-encode, langsung fix ID cols

# Kembalikan ID cols dari freq asli
df_imputed['policy_id'] = new_freq['policy_id'].values
df_imputed['worker_id'] = new_freq['worker_id'].values

# ── SPLIT KE 2 DATAFRAME ──────────────────────────────────────────────────────
freq_imputed = df_imputed[new_freq.columns].copy()

sev_imputed = new_sev.copy()
sev_imputed['injury_type']  = sev_imputed['injury_type'].fillna(sev_imputed['injury_type'].mode()[0])
sev_imputed['injury_cause'] = sev_imputed['injury_cause'].fillna(sev_imputed['injury_cause'].mode()[0])
sev_imputed['claim_length'] = sev_imputed['claim_length'].fillna(sev_imputed['claim_length'].median())

# ── FINAL CHECK ───────────────────────────────────────────────────────────────
print("=== freq_imputed ===")
print(f"Shape  : {freq_imputed.shape}")
print(f"Missing: {freq_imputed.isnull().sum().sum()}")
print(freq_imputed.dtypes)

print("\n=== sev_imputed ===")
print(f"Shape  : {sev_imputed.shape}")
print(f"Missing: {sev_imputed.isnull().sum().sum()}")
```

```python
# ── IMPUTE new_sev ──────────────────────────────────────────────────────────────

id_cols_sev = ['claim_id', 'policy_id', 'worker_id']

# Pisah numeric vs categorical
num_cols_sev = new_sev.drop(columns=id_cols_sev).select_dtypes(include='number').columns.tolist()
cat_cols_sev = new_sev.drop(columns=id_cols_sev).select_dtypes(include='object').columns.tolist()

# Numeric → MICE
df_imp_sev = new_sev[num_cols_sev].copy().astype('float64')

imputer_sev = IterativeImputer(estimator=BayesianRidge(), max_iter=20, random_state=42, verbose=2)
new_sev[num_cols_sev] = imputer_sev.fit_transform(df_imp_sev)

# Categorical → modus
for col in cat_cols_sev:
    modus = new_sev[col].mode()[0]
    new_sev[col] = new_sev[col].fillna(modus)

# Validasi
print("Missing new_sev setelah imputasi:")
print(new_sev.isnull().sum()[new_sev.isnull().sum() > 0])
print(f"\nTotal missing: {new_sev.isnull().sum().sum()}")
```
