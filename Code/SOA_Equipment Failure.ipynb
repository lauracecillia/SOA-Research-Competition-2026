```python
import pandas as pd

path = "srcsc-2026-claims-equipment-failure.xlsx"

dfs = pd.read_excel(path, sheet_name=["freq", "sev"])
df_freq = dfs["freq"]
df_sev  = dfs["sev"]

df_freq.head(), df_sev.head()
```

```python
missing_freq = pd.DataFrame({
    "Missing_Count": df_freq.isna().sum(),
    "Missing_Percent": (df_freq.isna().sum() / len(df_freq))*100
})

missing_freq
```

```python
missing_sev = pd.DataFrame({
    "Missing_Count": df_sev.isna().sum(),
    "Missing_Percent": (df_sev.isna().sum() / len(df_sev)) * 100
})

missing_sev
```

```python
# Cek jumlah sebelum drop
print("Jumlah baris sebelum drop:", len(df_sev))

# Drop baris dengan exposure missing
df_sev = df_sev.dropna(subset=["exposure"])

# Cek jumlah setelah drop
print("Jumlah baris setelah drop:", len(df_sev))
```
