```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# ── Data Historis ────────────────────────────────────────
INFLATION_HISTORY = {
    2160: 0.0377, 2161: 0.0232, 2162: 0.0148, 2163: 0.0108,
    2164: 0.0022, 2165: 0.0071, 2166: 0.0155, 2167: 0.0216,
    2168: 0.0181, 2169: 0.0073, 2170: 0.0294, 2171: 0.0708,
    2172: 0.0598, 2173: 0.0276, 2174: 0.0239,
}

# ── 1. Estimasi Parameter Model (Calibration) ────────────
years = sorted(INFLATION_HISTORY.keys())
values = [INFLATION_HISTORY[y] for y in years]

# dI = Inf_t - Inf_{t-1}
# I_prev = Inf_{t-1}
X = np.array(values[:-1]).reshape(-1, 1)
y = np.array(np.diff(values))

model = LinearRegression().fit(X, y)
speed_fitted = -model.coef_[0]
target_fitted = model.intercept_ / speed_fitted
residuals = y - model.predict(X)
volatility = np.std(residuals)

print(f"--- Model Calibration Results ---")
print(f"Long-run Target: {target_fitted:.4%}")
print(f"Reversion Speed: {speed_fitted:.4f}")
print(f"Volatility (σ):  {volatility:.4%}\n")

# ── 2. Proyeksi 16 Tahun (Sampai 2190) ───────────────────
forecast_vasicek = {}
current = values[-1]
last_year = years[-1]
n_projection = 16 # Ditambah 1 tahun (15 + 1)

np.random.seed(42) # Konsistensi hasil

# Coba turunkan floor ke 0.1% agar tidak terlihat "nyangkut" di 0.5%
for i in range(1, n_projection + 1):
    year = last_year + i
    shock = np.random.normal(0, volatility)
    current = current + speed_fitted * (target_fitted - current) + shock

    # Biarkan lebih fleksibel
    forecast_vasicek[year] = max(0.001, current)


# ── 3. Output Hasil ──────────────────────────────────────
print(f"{'Year':<6} | {'Inflation':>10} | {'Status'}")
print("-" * 35)
for year, val in sorted(forecast_vasicek.items()):
    status = "New Extension" if year == 2190 else "Projected"
    print(f"{year:<6} | {val:>10.4%} | {status}")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# ── Data Historis ────────────────────────────────────────
INFLATION_HISTORY = {
    2160: 0.0377, 2161: 0.0232, 2162: 0.0148, 2163: 0.0108,
    2164: 0.0022, 2165: 0.0071, 2166: 0.0155, 2167: 0.0216,
    2168: 0.0181, 2169: 0.0073, 2170: 0.0294, 2171: 0.0708,
    2172: 0.0598, 2173: 0.0276, 2174: 0.0239,
}

# ── 1. Calibration (Sama seperti sebelumnya) ────────────
years = sorted(INFLATION_HISTORY.keys())
values = [INFLATION_HISTORY[y] for y in years]
X = np.array(values[:-1]).reshape(-1, 1)
y = np.array(np.diff(values))

model = LinearRegression().fit(X, y)
speed_fitted = -model.coef_[0]
target_fitted = model.intercept_ / speed_fitted
volatility = np.std(y - model.predict(X))

# ── 2. Monte Carlo Simulation (10.000 Jalur) ─────────────
n_projection = 16
n_simulations = 1000
last_val = values[-1]

# Matriks untuk menyimpan semua hasil simulasi
# Baris = Tahun, Kolom = Jalur simulasi
all_paths = np.zeros((n_projection, n_simulations))

np.random.seed(42)

for s in range(n_simulations):
    current = last_val
    for t in range(n_projection):
        shock = np.random.normal(0, volatility)
        current = current + speed_fitted * (target_fitted - current) + shock
        all_paths[t, s] = max(0.001, current) # Floor 0.1%

# Ambil Rata-rata dari 10.000 jalur (Hasil yang paling stabil)
mean_forecast = np.mean(all_paths, axis=1)
# Ambil Percentile (untuk melihat skenario terburuk/terbaik)
upper_bound = np.quantile(all_paths, 0.95, axis=1) # Skenario inflasi tinggi
lower_bound = np.quantile(all_paths, 0.05, axis=1) # Skenario inflasi rendah

# ── 3. Simpan ke DataFrame & Excel ───────────────────────
df_results = pd.DataFrame({
    "Year": range(years[-1] + 1, years[-1] + n_projection + 1),
    "Expected_Inflation": mean_forecast,
    "High_Scenario_95th": upper_bound,
    "Low_Scenario_5th": lower_bound
})

print("--- Hasil Proyeksi Monte Carlo (Stabil) ---")
print(df_results.to_string(index=False, formatters={'Expected_Inflation': '{:.4%}'.format,
                                                   'High_Scenario_95th': '{:.4%}'.format,
                                                   'Low_Scenario_5th': '{:.4%}'.format}))

```
